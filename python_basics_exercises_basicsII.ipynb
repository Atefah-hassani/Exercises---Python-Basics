### Check if a given year is a leap year
- https://pynative.com/python-basic-exercise-for-beginners/#h-exercise-18-check-if-a-given-year-is-a-leap-year

```python
def is_leap_year(year):
    # First check divisibility by 400
    if year % 400 == 0:
        return True
    # Then check divisibility by 100
    elif year % 100 == 0:
        return False
    # Then check divisibility by 4
    elif year % 4 == 0:
        return True
    # Otherwise, not a leap year
    else:
        return False

# Test cases
year1 = 2020
print(is_leap_year(year1))  

year2 = 2025
print(is_leap_year(year2))  
```

### Print the following pattern
- https://pynative.com/python-if-else-and-for-loop-exercise-with-solutions/#h-exercise-2-print-the-following-pattern

```python
def print_pattern():
    rows = 5  # total number of rows
    
    for i in range(rows):          # outer loop runs 5 times
        for j in range(1, i + 2):  # inner loop runs i+1 times
            print(j, end=" ")
        print()  # print a new line after each row


# Call the function
print_pattern()
```

### Write a program to count occurrences of all characters within a string
- https://pynative.com/python-string-exercise/#h-exercise-10-write-a-program-to-count-occurrences-of-all-characters-within-a-string

```python
def count_characters(s):
    result = {}
    for char in s:
        result[char] = s.count(char)
    return result

# Test
str1 = "Apple"
print(count_characters(str1))
```

### Check if a value exists in a dictionary
- https://pynative.com/python-dictionary-exercise-with-solutions/#h-exercise-13-check-if-a-value-exists-in-a-dictionary

```python
def check_value_in_dict(d, value):
    # Convert dictionary values to a list
    values_list = list(d.values())
    
    # Check if the value exists
    if value in values_list:
        print(f"{value} present in a dict")
    else:
        print(f"{value} not present in a dict")


# Test the function
sample_dict = {'a': 100, 'b': 200, 'c': 300}
check_value_in_dict(sample_dict, 200)
```
