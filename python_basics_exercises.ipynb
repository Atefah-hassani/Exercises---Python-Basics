### 1. Print characters present at an even index number
 - https://pynative.com/python-basic-exercise-for-beginners/#h-exercise-3-print-characters-present-at-an-even-index-number

```python
# Accept a string from the user
s = input("Enter a string: ")
```

```python


# Calculate the length of the string
length = len(s)

# Iterate through even index numbers
for i in range(0, length, 2):
    print(s[i])
```

### 3. Merge two lists using the following condition
https://pynative.com/python-basic-exercise-for-beginners/#h-exercise-10-merge-two-lists-using-the-following-condition

```python
list1 = [10, 20, 25, 30, 35]
list2 = [40, 45, 60, 75, 90]
```

```python
result_list = []
```

```python
for item in list1:
   if item % 2 != 0:
       result_list.append(item)
```

```python
for item in list2:
    if item % 2 == 0: 
        result_list.append(item)
```

```python
print (result_list)
```

### 8. Remove Duplicates from list
https://pynative.com/python-list-exercise-with-solutions/#h-exercise-12-remove-duplicates-from-list

```python
list_with_duplicates = [1, 2, 2, 3, 1, 4, 5, 4]

unique_elements_set = set(list_with_duplicates)


unique_list = list(unique_elements_set)
print(f"Original list: {list_with_duplicates}")
print(f"List with unique elements: {unique_list}")
```

### 9. Delete a list of keys from a dictionary
https://pynative.com/python-dictionary-exercise-with-solutions/#h-exercise-12-delete-a-list-of-keys-from-a-dictionary

```python
sample_dict = {
    "name": "Kelly",
    "age": 25,
    "salary": 8000,
    "city": "New york"
}

# Keys to remove
keys = ["name", "salary"]
```

```python
for i in keys:
    sample_dict.pop(i)
print(sample_dict)
```
